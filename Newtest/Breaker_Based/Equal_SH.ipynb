```python
import yfinance as yf
import pandas as pd
import numpy as np
import requests
import mplfinance as mpf
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from lightweight_charts import Chart
```

```python
#Objective - To detect equal swing highs to identify liquidity 
```

```python
Ticker = 'SAIL.NS'

BTC = yf.Ticker(Ticker).history(period='100d', interval='1h' )
if BTC.empty:
     print(f'No data found for {Ticker}')



BTC["Swing_High"] = (BTC['High']>BTC['High'].shift(-1))&(BTC['High']>BTC['High'].shift(1))
BTC["Swing_Low"] = (BTC['Low']<BTC['Low'].shift(-1))&(BTC['Low']<BTC['Low'].shift(1))

BTC["Swing_High_value"] = np.where(BTC["Swing_High"], BTC["High"], np.nan)
BTC["Swing_High_value"] = BTC["Swing_High_value"].ffill()
BTC["Swing_Low_value"] = np.where(BTC["Swing_Low"], BTC["Low"], np.nan)
BTC["Swing_Low_value"] = BTC["Swing_Low_value"].ffill()

BTC["Swing_High_Return"] = (BTC['Swing_High_value'].shift(1)/BTC['Swing_High_value'])-1
BTC["Swing_Low_Return"] = (BTC['Swing_Low_value']/BTC['Swing_Low_value'].shift(1))-1


Conditions_ESH = [
    (BTC['Swing_High'] == True)&
    (BTC['Swing_High_Return'] < 0.001)&
    (BTC['Swing_High_Return'] > 0)
]

Label_ESH = ['Equal Swing High']

BTC['Equal_Swing_High'] = np.select(Conditions_ESH, Label_ESH, default='')

Conditions_ESL = [
    (BTC['Swing_Low'] == True)&
    (BTC['Swing_Low_Return'] < 0.001)&
    (BTC['Swing_Low_Return'] > 0)
]   
Label_ESL = ['Equal Swing Low']
BTC['Equal_Swing_Low'] = np.select(Conditions_ESL, Label_ESL, default='')


print(BTC)  
BTC.to_csv("SAILESL.csv")
```
